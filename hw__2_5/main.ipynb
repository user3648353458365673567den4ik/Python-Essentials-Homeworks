```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
petal_length = iris['petal_length']
```

```python
def show_petal_length_hist(iris_df):
    assert not iris_df["petal_length"].empty
    petal_length_hist = sns.histplot(data=iris_df["petal_length"])
    petal_length_hist.set_title("Распределение длины лепестков")
    
show_petal_length_hist(iris)
```

```python
def plot_box_data(iris_df, plot_min: bool = True, plot_max: bool = True, plot_median: bool = True):
    # проверка на соответсвие типов
    assert isinstance(plot_min, bool)
    assert isinstance(plot_max, bool)
    assert isinstance(plot_median, bool)

    assert not iris_df["petal_length"].empty
    assert not iris_df["petal_width"].empty
    
    
    iris_groups = iris_df.groupby('species')[["petal_length", "petal_width"]]  # создание перемонной, подходящей под задачу

    if plot_max:  # вывод графика относительно максимальных значений
        box_petal_max = sns.boxplot(data=iris_groups.max())
        box_petal_max.set_title("Распределение лепестков относительно максимальных длинн")
        plt.show()
    
    if plot_min:  #  вывод графика относительно минимальных значений
        box_petal_min = sns.boxplot(data=iris_groups.min())
        box_petal_min.set_title("Распределение лепестков относительно минимальных длинн")
        plt.show()
    
    if plot_median:  # вывод графика относительно медианных значений
        box_petal_median = sns.boxplot(data=iris_groups.median())
        box_petal_median.set_title("Распределение лепестков относительно медианной длины")
        plt.show()
        
plot_box_data(iris)  # допускается отключение графиков в аргументах
```

```python
def scatter_size_data(iris_df):
    assert not iris_df['petal_length'].empty
    assert not iris_df['petal_width'].empty
    
    scatter_petal_width_to_length = sns.scatterplot(x=iris_df['petal_length'], y=iris_df['petal_width'])
    scatter_petal_width_to_length.set_title("Отношение ширины лепестка к его длине")
    plt.show()
    
scatter_size_data(iris)
```
